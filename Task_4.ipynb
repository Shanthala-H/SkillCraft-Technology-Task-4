```python
import plotly.express as px

# Define campaign columns
campaign_cols = ['acceptedcmp1','acceptedcmp2','acceptedcmp3','acceptedcmp4','acceptedcmp5']

# Calculate total campaign accepted per customer
df['total_campaign_accepted'] = df[campaign_cols].sum(axis=1)

# Age distribution
fig = px.histogram(df, x='age', nbins=20, title='Age Distribution')
fig.show()

# Income vs Total Spending
fig = px.scatter(df, x='income', y='total_spent', color='age', size='total_spent',
                 title='Income vs Total Spending')
fig.show()

# Campaign Responses
fig = px.bar(df, x=df.index, y='total_campaign_accepted',
             title='Total Campaign Responses per Customer')
fig.show()
```

```python
# Step 9: Apply colors

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Total Spending Histogram - Green
sns.histplot(df['total_spent'], bins=20, color='green')
plt.title("Total Spending Distribution")
plt.show()

# Age Distribution - Blue
sns.histplot(df['age'], bins=20, color='blue')
plt.title("Age Distribution")
plt.show()

# Campaign Response - Red / Green
sns.countplot(x='response', data=df, palette=['red','green'])
plt.title("Campaign Response")
plt.show()

# Plotly Interactive - Income vs Total Spending
fig = px.scatter(df, x='income', y='total_spent', color='age', size='total_spent',
                 color_continuous_scale=px.colors.sequential.Viridis,
                 title='Income vs Total Spending')
fig.show()
```
